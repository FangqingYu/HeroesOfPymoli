## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#purchase_data.head()
```

```python
bins = [purchase_data['Age'].min()-1, 9, 14, 19, 24, 29, 34, 39, purchase_data['Age'].max()+1]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
by_Age_Range = purchase_data.groupby(pd.cut(purchase_data['Age'], bins, labels = group_names))
```

```python
by_Age_Range['SN'].nunique()
```

```python
#constract table from num of unique players by age range
Player_by_Age = pd.DataFrame(by_Age_Range['SN'].nunique()).rename(columns = {'SN': 'Total Count'})

#Calculate and add column 'Percentage of Players'
Player_by_Age['Percentage of Players'] = 100*(Player_by_Age['Total Count'] / Player_by_Age['Total Count'].sum())
Player_by_Age['Percentage of Players'] = Player_by_Age['Percentage of Players'].map('{:.2f}'.format)

Player_by_Age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting

```python
df1 = pd.DataFrame(purchase_data.groupby('SN').count()['Purchase ID']).rename(columns={'Purchase ID': 'Purchase Count'})
df2 = pd.DataFrame(purchase_data.groupby('SN').mean()['Price']).rename(columns={'Price': 'Average Purchase Price'})
df3 = pd.DataFrame(purchase_data.groupby('SN').sum()['Price']).rename(columns={'Price': 'Total Purchase Value'})
```

```python
Spender_Summary = pd.merge(df1, df2, how ='outer', on='SN')
Spender_Summary = pd.merge(Spender_Summary, df3, how='outer', on='SN').sort_values('Total Purchase Value', ascending = False)
```

```python
Spender_Summary['Average Purchase Price'] = Spender_Summary['Average Purchase Price'].map('${:,.2f}'.format)
Spender_Summary['Total Purchase Value'] = Spender_Summary['Total Purchase Value'].map('${:,.2f}'.format)

Spender_Summary.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
df_count = pd.DataFrame(purchase_data.groupby('Item ID').count()['Purchase ID']).rename(columns={'Purchase ID': 'Purchase Count'})

df_total_value = pd.DataFrame(purchase_data.groupby('Item ID').sum()['Price']).rename(columns={'Price': 'Total Purchase Value'})

# drop duplicates to get a price menu
df_item_price = purchase_data[['Item ID','Item Name', 'Price']].drop_duplicates().set_index('Item ID')

# merge count, sum of prices and price on Item ID
best_seller = pd.merge(df_item_price, df_count, on='Item ID')
best_seller = pd.merge(best_seller, df_total_value, on='Item ID')

# rename column and sort
# rearrange columns
best_seller = best_seller.rename(columns={'Price': 'Item Price'}).sort_values('Purchase Count', ascending=False)
best_seller = best_seller[['Item Name','Purchase Count','Item Price', 'Total Purchase Value']]

# copy for formatting
best_seller_f = best_seller

best_seller_f['Item Price'] = best_seller_f['Item Price'].map('${:,.2f}'.format)
best_seller_f['Total Purchase Value'] = best_seller_f['Total Purchase Value'].map('${:,.2f}'.format)

best_seller_f.head(10)
```
